```python
from paddleocr import PaddleOCR
import cv2
import matplotlib.pyplot as plt

# Initialize the PaddleOCR model with Vietnamese language support
ocr = PaddleOCR(use_angle_cls=True, lang='vi')

# Load the image
image_path = 'cccd.jpg'  # Replace with your actual image path
img = cv2.imread(image_path)

# Perform OCR on the image
result = ocr.ocr(image_path)

# Extract and print the text from the image
for line in result:
    for word_info in line:
        word = word_info[1][0]  # Extract the recognized word
        print(word)

# Display the image with matplotlib
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Input Image')
plt.show()
```
